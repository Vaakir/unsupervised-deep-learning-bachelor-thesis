```python
import requests
import os
import subprocess

# GitHub personal access token and username
GITHUB_TOKEN = '' # WHAT TOKEN EXACTLY? : https://youtu.be/iLrywUfs7yU?t=64
GITHUB_USERNAME = '' # YOUR USERNAME

# Headers for GitHub API authentication
headers = {
    'Authorization': f'token {GITHUB_TOKEN}',
    'Accept': 'application/vnd.github.v3+json'
}

def search_repositories(query, per_page=10, max_repos=50, print_progress=True):
    url = f'https://api.github.com/search/repositories?q={query}&per_page={per_page}'
    repositories = []
    
    page = 1
    while len(repositories) < max_repos:
        if print_progress:
            print(len(repositories), max_repos, page)
        response = requests.get(url + f"&page={page}", headers=headers)
        if response.status_code == 200:
            data = response.json()
            repositories.extend(data['items'])
            if len(data['items']) < per_page:
                break  # No more repositories
        else:
            print(f"Error fetching repositories: {response.status_code}")
            break
        page += 1
    return repositories[:max_repos]


repositories = []
search_terms = ['mri autoencoder']

for term in search_terms:
    repositories.extend(search_repositories(term, per_page=1_00, max_repos=10_000))

# Remove duplicates by repository id
unique_repositories = {repo['id']: repo for repo in repositories}.values()
print(f"unique_repositories: {len(unique_repositories)}")

def clone_repositories(repositories, download_path='repositories'):
    "Clone repositories to local directory"
    if not os.path.exists(download_path):
        os.makedirs(download_path)
    
    # for index, repo in enumerate(repositories):
    lis = list(repositories)
    for index in range(0,len(lis)):
        repo = lis[index]
        repo_name = f"{index}_{repo['name']}"
        clone_url = repo['clone_url']
        repo_path = os.path.join(download_path, repo_name)
        
        if not os.path.exists(repo_path):
            print(f"Cloning {repo_name}...")
            subprocess.run(['git', 'clone', clone_url, repo_path])
        else:
            print(f"{repo_name} already cloned.")

clone_repositories(unique_repositories)
```
