```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x0,y0 = -1,-1
x1,y1 = 1,1
r = 2

plt.figure(figsize=(5,5))

X = []
Y = []
for angle in np.linspace(0, np.pi*2):
    s = np.sin(angle)
    c = np.cos(angle)
    p = 0.5 + 0.5*(s*(x1-x0) + c*(y1-y0)) / ((x1-x0)**2+(y1-y0)**2)**0.5
    X.append(x0+(x1-x0)*p + s*r)
    Y.append(y0+(y1-y0)*p + c*r)
plt.plot(X,Y)

plt.scatter([-5,-5,5,5],[-5,5,-5,5],c="white")

# THE "PROPER" WAY TO DO IT

# Ellipse parameters
x_center, y_center = 0, 0  # Center of the ellipse
a, b = 3.4, 2  # Semi-major and semi-minor axes
phi = np.pi / 4  # Rotation angle in radians (45 degrees here)

# Generating ellipse points
theta = np.linspace(0, 2 * np.pi, 500)  # 500 points for smoothness
X = a * np.cos(theta)  # Unrotated x-coordinates
Y = b * np.sin(theta)  # Unrotated y-coordinates

# Applying rotation
X_rot = x_center + X * np.cos(phi) - Y * np.sin(phi)
Y_rot = y_center + X * np.sin(phi) + Y * np.cos(phi)

# Plot the rotated ellipse
plt.plot(X_rot, Y_rot, label=f"Rotated Ellipse (phi={phi:.2f} rad)")
plt.gca().set_aspect('equal', adjustable='box')  # Ensure aspect ratio is equal
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)  # x-axis
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)  # y-axis
```
