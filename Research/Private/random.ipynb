```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def f(x):
    return np.sin(x)

x = np.linspace(0, 2 * np.pi, 100)
y = f(x)
plt.plot(x, y)
```

```python
random = np.random.uniform(0,2*np.pi,10)

random_y = f(random)

plt.scatter(random, random_y, c='r')
```

```python
random[np.argmin(random_y)]
```

```python
x = 3

# Use the gradient descent algorithm to minimize f(x)

def gradient_descent(x, learning_rate=0.01, iterations=1000):
    for _ in range(iterations):
        gradient = 2 * (f(x) - np.sin(x))
        x -= learning_rate * gradient
    return x

optimal_x = gradient_descent(x)
```
